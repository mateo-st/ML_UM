### __Bayesian Inference__

Types of inference:
- Deductive: an argument where the conclusion is necessarily inferred from the premises. Conclusions are safe to be drawn: general -> specific.
- Inductive: study of the tests that allow asking for the probability of the assumptions: specific -> general. 

We must include probability in our thinking process. Deductive reasoning means any given proposition is considered true or false, without admitting any other value between the two. Approximate reasoning provides theoretical models that simulate the ability to reason under conditions of uncertainty, when the truthfulness of a statement or hypothesis is not known with absolute certainty: eg. Bayesian reasoning (based on Bayes' theorem).

- Frequentist Probability: the value to which the probability of an event tends when when an event is repeated.
- Bayesian Probability: subjectively interprets probability and uses it to express belief regarding an afirmation given certain evidence. New information can be incorporated as to formulate again the probability.

Probability is a tool to measure our __beliefs__.

____Uncertainty:____
Every inference statement we make must deal with uncertainty.Types of uncertainty:
- Ignorance: the limits of our knowledge lead us to be ignorant about many things.
- Randomness: the influence of chance.
- Vagueness: many of the concept that we use in our thinking have a certain degree of subjectivity?

Bayesian inference is the philosophy which states that to understand human opinion as it is, limited by ignorance and uncertainty, we must use the __measurement of probability__ as the most important tool to represent the strength of our beliefs.

1. __Pre-test probability:__ We have a preconceived idea of the probability of an event happening.
2. __Post-test probability:__ New information reaches us, updating the previously held notions.

These calculations are made through ___Bayes Theorem___ which incorporates both the new data and the pre-test probability.

Pre-test Probability 	$\rightarrow$   New information 	$\rightarrow$   Application of Bayes' Theorem	$\rightarrow$   Updated Probability

Post-test Probability $\displaystyle \propto$ Pre-test Probability

The difference between classical (frequentist) and Bayesian inference is that the latteer considers "prior knowledge" while frequentist only what emerges from the experiment.

Does conditional probability imply causality?

If there is a causal relationship between events A and B:
- P(A $\lvert$ B) $\neq$ P(A)
- In particular, if it is a strict cause, P(A $\lvert$ B) = 1
    - However, a statistical correlation can exist without causality.

### Classification problems

In machine learning, quantitative data and responses are under the term "regression analysis". Qualitative data however, is assimilated to __classification problems__.

In qualitative problems, we work with _features_:
- Age, gender, salary, employment.
- From the given set of features, the idea is to predict the behaviour of a subject.
- Example: a bank wants to know whether the given person can afford to pay a loan.

Feature engineering is key when working with datasets. Which attributes are important for the model.

Conceptually, a classification method is a function that divides space into two classes (binary classificator). Some methods can also define a security threshold. A sentiment classificator can, for example, classify a text into (positive, neutral, negative). 

Probabilistic and non-probabilistic classifiers differ in the type of output they produce. In the first case, the output is the probability that a new object belongs to a certain class. Thus, given a text, a sentiment classificator's output is:

$ Output = \begin{bmatrix}
Positive,     p=0.8\\
Negative,   p=0.1 \\
Neutral,     p=0.1\\
\end{bmatrix} $



#### Bayesian classifier

Takes it name from the Bayes equation. Given a set of training points of C classes, the goal is to be able to determine the predictive probabilities for each of the potential classes C. Being x a given data point, and T the target class:

P($T_{new}$ = $c$ $\lvert$ $x_{new}$, X, t) , meaning, we want to find the probability of $x_{new}$ being a subject of $T_{new}$ class. X y t are previous data and their belonging (or not) to the class. 

In the example of the loan bank, X are previous' clients data, t is their loan payment history (whether these previous clients payed or not), $x_{new}$ is the new client's information, and $T_{new}$ is the prediction of whether the new client will pay or not.

${\dfrac{P(Ti\lvert X) = P(X \lvert T_i)* P(T_i)}{P(X)}}$

$P(T_i)$ can be calculated in two ways:
- Assume constant probability for all classes
    - $P(T_1)$ = $P(T_2)$ = ... = $P(T_n)$ = $1/n$
- Estimate probability with latest data
    - $P(T_i)$ = $C_i/D$ ; where $C_i$ is the amount of time the class appears in the dataset, and $D$ the volume of data.

### Metrics

Supervised learning implies we use part of the data to train the model and part to test (80/20).
How "good" is the model we are applying? The scientific community shares the same metrics. Not one given metric represents the fitness of the model, different .

___Confusion matrix___

Matrix that represents accurate predictions (true positives and true negatives), false negatives and false positives.
- TP: predicted class is 1, real class is 1.
- TN: predicted class is 0, real class is 0.
- FP: predicted class is 1, real class is 0.
- FN: predicted class is 0, real class is 1.

__Accuracy__

${\dfrac {(TP + TN)}{(TP + FP + TN + FN)} }$


__Precision__

Useful for: Spam Mail

${\dfrac {(TP)}{(TP + FP)} }$ 


__Recall__

Useful for: Terrorist Detection

${\dfrac {(TP)}{(TP + FN)} }$ 


__F1 Score__ 

Useful for: Finding a balance between precision and recall; harmonic mean.

${\dfrac {(2 * Precision * Recall)}{(Precision + Recall)} }$ 

__ROC Curve__

The larger the area, the better the model.

$False Positive Rate = {\dfrac {(FP)}{(TN + FP)} }$ 

$True Positive Rate = {\dfrac {(TP)}{(TP + FN)} } = Recall$

#### Logistic regression

__Threshold:__ a dividing value that separates attributes into two binary values (tumor size vs. malignant tumor?).

Logistic regression transforms the threshold into a sigmoidal probability function.

$ p(x) = {\dfrac {1}{1 + e^{-\dfrac{(x - \mu)}{s}} }}$ 

where $ \mu $ = location parameter, $ s $ = scale parameter

This is the same as

$ p(x) = {\dfrac {1}{1 + e^{-b_1 + b_2} }}$ 

We define _maximum likelihood curve_ as the sigmoidal function that maximizes the probability of getting the training dataset.

We generalize the loss function we saw in least squares:

$\frac{1}{n} \sum_{n=1}^{n} (y_m - y_d)^2$

E.g.: Define whether a tumor is malignant.
- y = 1, P must be high (low probability should be penalized)
- y = 0, P must be low (high probability penalty)

### PCA: __Principal Component Analysis__

Cuando tenemos un dataset con muchas columnas, muchos features, la técnica de PCA nos permite resumir el conjunto de variables en un número menor de variables. Un número menor de variables, representativas, pueden explicar la mayor parte de la variabilidad de los datos haciendo más simple la manipulación y entendimiento de los datos. El desafío de fondo es lograr minimizar la cantidad de atributor de forma de perder la menor cantidad de información posible.

Existen features que no aportan información y existen features redundantes con otros features.

La matriz de covarianza mide qué tan disperso están los datos.

$ M = \begin{bmatrix}
Var(x), Cov (X,Y)\\
Cov (Y,X), Var(Y) \\
\end{bmatrix} $

Si realizamos una transformación lineal de un vector cualquiera por la matriz de covarianza, estira el plano en 2 direcciones perpendiculares (estas direcciones son los vectores propios). El vector propio más importante es el que tiene el valor propio más grande.
